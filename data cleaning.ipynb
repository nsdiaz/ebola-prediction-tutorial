```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('sl_ebola.csv')
df.shape
```

```python
df.head(10)
```

```python
df.tail()
```

```python
df.info()
```

```python
# Understanding the value count of certain features
print(df.Country.value_counts(), '\n')
print(df.District.value_counts(), '\n')
print(df['Ebola measure'].value_counts(), '\n')
print(df['Ebola data source'].value_counts(), '\n')
print(df['Epi week'].value_counts(), '\n')
print(df['Indicator type'].value_counts(), '\n')
```

```python
# Dropping features with too many missing values or irrelevant for our analysis
df.drop(columns=['Age group', 'Gender', 'Data as of', 'Outbreak identifier', 'Country', 'Indicator type'], inplace = True)
df.head()
```

```python
df2 = df[df.District != 'Aggregate']
df2.head(10)
```

```python
df3 = pd.pivot_table(df2, values=['Numeric'], index=['District', 'Epi week', 'Ebola data source'], columns=['Case definition'], aggfunc=np.sum)
df3.head()
```

```python
df4 = pd.DataFrame(df3.to_records()) 
df4.rename({'(\'Numeric\', \'Confirmed\')' :  'Confirmed_cases', '(\'Numeric\', \'Probable\')' :  'Probable_cases', }, axis = 1, inplace = True)
df4.head(10)
```

```python
df4.shape
```

```python
df5 = pd.pivot_table(df4, values=['Confirmed_cases', 'Probable_cases'], index=['District', 'Epi week'], columns=['Ebola data source'], aggfunc=np.sum)
df6 = pd.DataFrame(df5.to_records()) 
df6.rename({'(\'Confirmed_cases\', \'Patient database\')' :  'patient_database_confirmed', '(\'Confirmed_cases\', \'Situation report\')' :  'situation_report_confirmed', '(\'Probable_cases\', \'Patient database\')' :  'patient_database_probable', '(\'Probable_cases\', \'Situation report\')' :  'situation_report_probable', }, axis = 1, inplace = True)
```
